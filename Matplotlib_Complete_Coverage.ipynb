## 📺 Episode 3: Matplotlib Architecture – The Hierarchy of Plots

```python

fig = plt.figure(figsize=(8, 6), facecolor='lightgray') # 8x6 inch Figure, light gray background
plt.title("Yeh Figure Title Hai (Poora Canvas)") # Pyplot style title for the entire figure
plt.show()

# You can also print Figure objects to see them
print(fig)
```

```python
x=np.linspace(0, 5)
y=np.sin(np.linspace(0, 5))
fig = plt.figure(figsize=(10, 5))

# [left, bottom, width, height] as fractions of Figure size
ax1 = fig.add_axes([0.1, 0.1, 0.4, 0.8]) # Left Axes
ax2 = fig.add_axes([0.55, 0.1, 0.4, 0.8]) # Right Axes

ax1.plot(x, y, color='blue',linestyle='--')
ax1.set_title("Sine Wave (Ax1)") # Notice ax.set_title()
ax1.set_xlabel("Time (s)")
ax1.set_ylabel("Amplitude")

ax2.bar(['A', 'B', 'C'], [3, 7, 2], color='green')
ax2.set_title("Bar Chart (Ax2)")
ax2.set_xlabel("Categories")
ax2.set_ylabel("Count")

fig.suptitle("Figure with Two Axes (add_axes)") # Overall Figure title
plt.show()
```

```python

x = np.linspace(0, 10, 20)
y1 = x**2
y2 = x**1.5
plt.plot(x, y1, label='Quadratic') # Label for legend
plt.plot(x, y2, label='Power 1.5')

# Custom ticks
plt.xticks([0, 2, 4, 6, 8, 10], ['Start', '2', '4', '6', '8', 'End'])
plt.yticks([0, 25, 50, 75, 100],['bottom', 25, 50, 75, 'Top'])

plt.xlabel("Mera X-axis Label")
plt.ylabel("Mera Y-axis Label")
plt.title("Ticks, Labels, aur Legend Ke Saath Plot")

plt.legend(loc='upper right', shadow=True) # Show legend at upper left
plt.grid(True, linestyle='--', alpha=0.1) # Add grid lines
plt.show()
```

```python

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8)) # 2 rows, 2 columns

# Access Axes using indexing
axes[0, 0].plot(np.random.rand(10), color='red')
axes[0, 0].set_title("Top Left Plot")


axes[1, 1].bar(['X', 'Y'], [5, 9], color='purple')
axes[1, 1].set_title("Bottom Right Plot")

plt.tight_layout() # Adjust layout for better spacing
plt.show()
```

Deep Dive into plt.plot() & Customization
First, plt.plot() is for line charts, and it's very versatile.

```python


plt.figure(figsize=(7, 4)) # You can also set figure size in Pyplot
x = np.linspace(0, 10, 50)
y = np.sin(x)
plt.plot(x, y,
         label='Sine Wave',     # Label for legend
         color='red',           # Line color
         linestyle='--',        # Line style (dashed)
         marker='o',            # Marker type (circle)
         markersize=5,          # Marker size
         linewidth=2)           # Line thickness 
plt.title("plt.plot() ka istemal karke Sine Wave", fontsize=14)
plt.xlabel("X-axis (Time)", fontsize=12)
plt.ylabel("Y-axis (Amplitude)", fontsize=12)
plt.legend(loc='upper right') # Legend position
plt.grid(True, linestyle=':', alpha=0.6) # Grid with dotted lines
plt.show()
```

```python
plt.figure(figsize=(7, 4))
categories = ['Apples', 'Bananas', 'Oranges', 'Grapes']
values = [25, 40, 30, 15]
plt.bar(categories, values, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'], edgecolor='black')
plt.title("Fruit Sales using plt.bar()")
plt.xlabel("Fruit Type")
plt.ylabel("Sales Count")
plt.show()
```

```python
plt.figure(figsize=(7, 4))
plt.barh(categories, values, color='teal') # barh for horizontal
plt.xlabel("Sales Count")
plt.ylabel("Fruit Type")
plt.title("Horizontal Bar Chart")
plt.show()
plt.scatter(categories,values)
```

```python
plt.figure(figsize=(7, 5))
np.random.seed(42) # for reproducibility
x_data = np.random.rand(50) * 10
y_data = np.random.rand(50) * 10 + np.random.rand(50) * 2
sizes = np.random.rand(50) * 800 + 100 # Marker sizes based on random value
colors = np.random.rand(50) # Marker colors based on random value
plt.scatter(x_data, y_data,
            s=sizes,          # Size of markers
            c=colors,         # Color of markers
            cmap='viridis',   # Colormap to use
            alpha=0.7,        # Transparency
            edgecolor='black')# Marker edge color
plt.colorbar(label="Random Value") # Add a color bar 
plt.title("plt.scatter() ka istemal karke Scatter Plot")
plt.xlabel("X-Value")
plt.ylabel("Y-Value")
plt.show()
```

```python
# plt.figure(figsize=(7, 4))
data_for_hist = np.random.randn(1000) # 1000 random numbers (normal distribution)
plt.hist(data_for_hist,
         bins=30,             # Number of bins/bars
         color='darkcyan',    # Bar color
         edgecolor='black',   # Bar edge color
         alpha=0.8,           # Transparency
         density=False)       # True for probability density, False for counts
plt.title("plt.hist() ka istemal karke Histogram (Normal Distribution)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.pie([10, 20, 30, 40], labels=['A', 'B', 'C', 'D'], autopct='%1.1f%%')
plt.title("Pie Chart Distribution")
plt.show()
```

```python
fig2, axes = plt.subplots(2, 2, figsize=(10, 8)) # 2 rows, 2 columns, 'axes' is 2D array

axes[0, 0].plot(np.linspace(0, 5), np.exp(np.linspace(0, 5)), color='red')
axes[0, 0].set_title("Exponential Growth")

axes[0, 1].hist(np.random.randn(1000), bins=30, color='orange', alpha=0.7)
axes[0, 1].set_title("Normal Distribution")

axes[1, 0].scatter(np.random.rand(50), np.random.rand(50), color='teal', alpha=0.6)
axes[1, 0].set_title("Random Scatter")
axes[1, 0].set_xlabel("X-Value")
axes[1, 0].set_ylabel("Y-Value")

axes[1, 1].pie([10, 20, 30, 40], labels=['A', 'B', 'C', 'D'], autopct='%1.1f%%')
axes[1, 1].set_title("Pie Chart Distribution")


fig2.suptitle("Four Subplots in a 2x2 Grid")
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python

fig, ax = plt.subplots(figsize=(9, 6))

x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x), label='sin(x)', color='blue', linewidth=2)
ax.plot(x, np.cos(x), label='cos(x)', color='red', linestyle='--', alpha=0.7)

# Setting properties using ax methods
ax.set_title("Sine and Cosine Waves", fontsize=16, color='darkblue')
ax.set_xlabel("Angle (radians)", fontsize=12)
ax.set_ylabel("Value", fontsize=12)

# Set axis limits
ax.set_xlim(0, 10)
ax.set_ylim(-1.2, 1.2)

# Add legend
ax.legend(loc='lower left', frameon=True, fancybox=True, shadow=True)

# Add grid
ax.grid(True, linestyle=':', alpha=0.6)

# Add text annotation 
ax.text(3, 0.8, 'Peak of Sine', fontsize=10, color='blue')
ax.annotate('Cosine Crosses Zero', xy=(np.pi/2, 0), xytext=(np.pi/2 + 2, 0.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            fontsize=10, color='black')

# Customize ticks
ax.set_xticks(np.arange(0, 11, 1))
ax.set_yticks([-1, -0.5, 0, 0.5, 1])

# Modify existing artist
line1 = ax.lines[0] # First line object
line1.set_dashes([2, 2, 10, 2]) # Dashed with long dashes
plt.show()
```

```python
# Example of a reusable plotting function (OO approach)
def plot_data_on_axes(ax_object, data, title, color):
    ax_object.plot(data, color=color)
    ax_object.set_title(title)
    ax_object.set_xlabel("Data Point")
    ax_object.set_ylabel("Value")
    ax_object.grid(True)

fig_reusable, (ax_a, ax_b) = plt.subplots(1, 2, figsize=(12, 5))
plot_data_on_axes(ax_a, np.random.rand(20), "Dataset A (OO Function)", 'red')
plot_data_on_axes(ax_b, np.random.randint(0, 10, 20), "Dataset B (OO Function)", 'blue')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(7, 5)) # Convenient way to get Figure and Axes
x = np.linspace(0, 2 * np.pi, 100)
ax.plot(x, np.cos(x))

print(f"Type of ax.xaxis: {type(ax.xaxis)}") # Show the type
print(f"Type of ax.yaxis: {type(ax.yaxis)}")

# Rotate X-axis labels
ax.xaxis.set_tick_params(labelrotation=45)
# Change Y-axis label color
ax.yaxis.label.set_color('red')

ax.set_title("Custom Axis Properties Ke Saath Plot")
ax.set_xlabel("Angle (radians)")
ax.set_ylabel("Cosine Value")

plt.show()
"Ab Ticks, Labels, aur Legends. Ticks chote marks hote hain. Labels woh numbers ya text jo ticks ke saath hote hain, aur axis ke naam bhi labels hain. Legends tab kaam aate hain jab multiple lines ho."
```
