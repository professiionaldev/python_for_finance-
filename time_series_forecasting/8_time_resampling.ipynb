## Resampling is usually applied to a date time index, which was the reason why we converted the Date column into a date time index.

### ðŸ“… Pandas `resample()` Frequency Aliases

| Code     | Description                                             |
|----------|---------------------------------------------------------|
| `B`      | Business day frequency                                  |
| `D`      | Calendar day frequency                                  |
| `W`      | Weekly frequency                                        |
| `M`      | Month end frequency                                     |
| `SM`     | Semi-month end frequency (15th and end of month)        |
| `BM`     | Business month end frequency                            |
| `CBM`    | Custom business month end frequency                     |
| `MS`     | Month start frequency                                   |
| `SMS`    | Semi-month start frequency (1st and 15th)               |
| `BMS`    | Business month start frequency                          |
| `CBMS`   | Custom business month start frequency                   |
| `Q`      | Quarter end frequency                                   |
| `BQ`     | Business quarter end frequency                          |
| `A` / `Y`| Year end frequency                                      |
| `BA`     | Business year end frequency                             |
| `AS`     | Year start frequency                                    |
| `BAS`    | Business year start frequency                           |
| `BH`     | Business hour frequency                                 |
| `H`      | Hourly frequency                                        |
| `T` / `min` | Minutely frequency                                   |
| `S`      | Secondly frequency                                      |
| `L` / `ms` | Millisecond frequency                                 |
| `U` / `us` | Microsecond frequency                                 |
| `N`      | Nanosecond frequency                                    |


```python
# data.resample(rule='M', how=None, axis=0, fill_method=None, closed=None, label=None,
#               convention='start', kind=None, loffset=None)
```

```python
# Calendar day frequency Means 
data.resample(rule='D').mean().head()
```

```python
data.resample(rule='W').mean().head()
```

```python
# Business month end frequency Means 
data.resample(rule='BM').mean().head()
```

```python
data.resample('Y').sum()
```

## We can also use various aggregate functions, such as max() , min(), or std(), as shown in the following code segment:
## yearly frequency max

```python
data.resample(rule='YE').max().head()
```

## The following code shows how we can apply the min() 
## yearly frequency min 

```python
data.resample(rule='A').min().head()
```

```python
#yearly frequency standard deviation 
data.resample(rule='A').std().head()
```

# Timeshifts

## tshift() (now deprecated) was used to shift the datetime index itself, not the data.

```python
dataframe = pd.read_csv('walmart_stock.csv',index_col='Date') 
dataframe.index = pd.to_datetime(dataframe.index)
```

```python
dataframe.head()
```

```python
dataframe.tail()
```

```python
dataframe.shift(periods=1).head()
```

```python
dataframe.shift(periods=-1).tail()
```

```python
# Shift everything forward one month 


# dataframe.tshift(periods=1,freq='ME').head()

# âŒ `tshift()` is deprecated in Pandas

# Use the new method instead:


dataframe.shift(periods=3, freq='M').head()


```
